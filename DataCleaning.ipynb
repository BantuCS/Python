```python
#DataCleaning: fixing bad data in your data set
#Bad Data: Empty Cells, Wrong Format, Wrong data,Duplicate cells
```

```python
import pandas as pd
df=pd.read_csv('dirtydata.csv')
print(df)
```

```python
#Cleaning empty cells: delete rows, replace empty values(fillna()) by zero
df1=df.fillna(0)
print(df1)
```

```python
#instead filling all empty cells as zero now i am fiiling with mean() of calories
avg=df["Calories"].mean() 
me=df["Calories"].median() 
mode=df["Calories"].mode() [0]
print(avg)
print(me)
print(mode)
df2=df.fillna(avg)
print(df2)
```

```python
#wrong format: convert data into correct formate
#pandas has to_datetime()
#df['Date'] = pd.to_datetime(df['Date'])
```

```python
#replacing wrong values
for x in df.index:
    if df.loc[x,"Duration"]>60:
        df.loc[x,"Duration"]=45
print(df)
```

```python
#checking duplicates are present
print(df.duplicated())
```

```python
#Removing duplicates
df.drop_duplicates(inplace= True)
print(df)
```
