```python
#Series is one dimensional array
import pandas as pd
name=['abc','def','pqr']
#creating a series by using list
ser=pd.Series(name)
```

```python
#displaying a series
print(ser)
```

```python
df=pd.read_csv("dirtydata.csv")
#creating a series by using csv
ser1=pd.Series(df['Duration'])
```

```python
#displaying series
print(ser1)
```

```python
#displaying first 5 rows in series
data=ser1.head(10)
print(data)
```

```python
#using loc function
data.loc[3:4]
```

```python
#using iloc function
data.iloc[8:10]
```

```python
#performing binary operations on series
#creating own index
name=['a','pqr','xyz','hjk','lpk','ramya']
ser3=pd.Series(name,index=['a','b','c','d','e','f'])
print(ser3)
```

```python
#performing binary operations on series
#adding two series 
ser2=pd.Series(df['Pulse'])
data3=ser2.head(6)
data2=ser1.head(6)
print(data2)
print(data3)
data2.add(ser2,fill_value=0)
```
